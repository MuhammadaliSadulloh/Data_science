```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sonlar=np.arange(10)
print(sonlar)
plt.plot(sonlar)
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,linewidth=2.0,linestyle='--')
plt.show()
```

```python
plt.plot(x,y,linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')
plt.show()
```

```python
plt.plot(x,y,linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')
plt.title('sin(x) funksiya grafigi')
plt.xlabel('pi')
plt.ylabel('sin(x)')
plt.grid()
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
sin=np.sin(x)
cos=np.cos(x)
plt.figure(figsize=(12,6))
plt.plot(x,sin,linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7)
plt.plot(x,cos,linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='o')
plt.title('sin(x) va cos(x) grafigi')
plt.xlabel('pi')
plt.legend(['sin(x)','cos(x)'],loc=0,fontsize='x-large')
plt.grid(color='tomato',alpha=0.5,linestyle='--')
plt.show()
```
