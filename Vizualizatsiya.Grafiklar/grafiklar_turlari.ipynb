```python

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 12, 15, 13, 17]

plt.plot(x, y)
plt.title('Chiziqli Grafik')
plt.xlabel('Vaqt')
plt.ylabel('Qiymat')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

df1 = pd.DataFrame({'X': [1, 2, 3, 4, 5],
                   'Y': [2, 4, 5, 4, 5]})

sns.scatterplot(x='X', y='Y', data=df1)
plt.title('Nuqtali Grafik')
plt.show()
```

```python
categories = ['A', 'B', 'C']
values = [10, 15, 7]

plt.bar(categories, values)
plt.title('Shtab Grafik')
plt.xlabel('Kategoriya')
plt.ylabel('Qiymat')
plt.show()
```
