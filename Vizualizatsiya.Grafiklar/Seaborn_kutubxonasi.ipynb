```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df=df[df.year>=2010] # 2010-yildan keyin chiqqan mashinalarni saqlab qolamiz
df=df[df.transmission!='Other'] # Faqat avtomat, mexanik va yarim-avtomat mashinalarni olib qolamiz
df=df[df.fuelType!='Other'] # Faqat benzin, dizel va gibrid mashinalarni olib qolamiz
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df,x='price')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df,x='price',height=5,aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df,x='mileage',height=5,aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df,x='mileage',hue='transmission',multiple='stack',height=5,aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df,x='mileage',col='transmission',height=5,aspect=1)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df,x='mileage',kind='kde',height=5,aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x='price', hue='fuelType', kind='kde', height=5, aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df,x='mileage',kde=True,height=5,aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
#kdeplot
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x='price', shade=True)
plt.xticks(rotation=90)
plt.show()
```

```python
#kdeplot
plt.figure(figsize=(10,5))
sns.kdeplot(data=df, x='price', hue='transmission', shade=True)
plt.xticks(rotation=90)
plt.show()
```

```python
# ikki o'lchamli KDE
sns.displot(data=df,x='price',y='year',kind='kde',height=5,aspect=2)
plt.show()
```

```python
sns.displot(data=df,x='mileage',kind='ecdf',height=5,aspect=2)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.displot(data=df, x="mileage", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both',color='c',linestyle='--',alpha=0.3) # setka (major, minor, both) qiymatlarini berib ko'ring
plt.minorticks_on()
plt.show()
```

```python
sns.displot(data=df, x="price", hue="fuelType", kind="ecdf", height=5, aspect=2)
plt.xticks(rotation=90)
plt.grid(which='both',color='c',linestyle='--',alpha=0.3) 
plt.minorticks_on()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.ecdfplot(data=df, x="price", hue="transmission")
plt.grid(which='both',color='c',linestyle='--',alpha=0.3) 
plt.minorticks_on()
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
sin=np.sin(x)
cos=np.cos(x)

plt.figure(figsize=(12,6))


plt.subplot(2,1,1)
plt.plot(x,sin,linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')
plt.title('sin(x) grafiki')
plt.grid(color='tomato',alpha=0.5,linestyle='--')


plt.subplot(2,1,2)
plt.plot(x,cos,linewidth=1.5,
         linestyle=':',
         alpha=0.7,
         marker='v')
plt.title('cos(x) grafiki')
plt.grid(color='tomato',alpha=0.5,linestyle='--')
plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/merc.csv")
df=df[df.year>=2010] # 2010-yildan keyin chiqqan mashinalarni saqlab qolamiz
df=df[df.transmission!='Other'] # Faqat avtomat, mexanik va yarim-avtomat mashinalarni olib qolamiz
df=df[df.fuelType!='Other'] # Faqat benzin, dizel va gibrid mashinalarni olib qolamiz
df.head()
```

```python
fig,ax=plt.subplots(2,1,figsize=(10,10))

fig.suptitle('Grafik 1')
sns.histplot(ax=ax[0],data=df,x='price')
sns.scatterplot(ax=ax[1],data=df,x='year',y='price',hue='fuelType',s=50)
plt.show()
```

```python
fig,ax=plt.subplots(2,2,figsize=(15,10))

fig.suptitle('Grafik 2')
sns.histplot(ax=ax[0,0],data=df,x='price')
sns.histplot(ax=ax[1,0],data=df,x='mileage')
sns.scatterplot(ax=ax[0,1],data=df,x='year',y='price',hue='fuelType',s=50)
sns.scatterplot(ax=ax[1,1],data=df,x='year',y='mileage',hue='fuelType',s=50)

# har bir grafik uchun nom 

ax[0,0].set_title('Price distrubution')
ax[1,0].set_title('Mileage distribution')
ax[0,1].set_title('Price vs Year correlarion')
ax[1,1].set_title('Mileage vs Year correlation')
plt.show()
```
