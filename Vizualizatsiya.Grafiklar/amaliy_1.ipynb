```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel('/content/drive/MyDrive/newcars.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# Histogramma
plt.figure(figsize=(8,5))
df['price'].hist(bins=20)
plt.title('Histogramma')
plt.xlabel('Qiymatlar')
plt.ylabel('Frekvensiya')
plt.show()
```

```python
# Scatterplot
plt.figure(figsize=(8,6))
sns.scatterplot(x='price', y='company', data=df)
plt.title('Scatterplot')
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(6,5))
sns.boxplot(data=df, y='price')
plt.title('Boxplot')
plt.show()
```
