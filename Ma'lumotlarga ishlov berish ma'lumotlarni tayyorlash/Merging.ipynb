```python
B = pd.DataFrame({'Transport': ['poezd', 'avtomabil', 'tramvay'], 
                  'Narxi':['98000', '10000', '65000']})
                  
B
```

```python
# MERGE
```

```python

# ICHKI UYG'UNLASHISH
```

```python
merged=pd.merge(A,B,how='inner')
```

```python
merged.head()
```

```python
merged_outer=pd.merge(A,B,how='outer')
```

```python
merged_outer.head()
```

```python
merged_left=pd.merge(A,B,how='left')
```

```python
merged_left
```
