```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/housing_data_08-02-2021.csv')
df
```

```python
import numpy as np
df['price'].replace('Договорная',np.nan,inplace=True)
df['size'].replace('Площадьземли:1сот',100,inplace=True)
df['price']=df['price'].astype(float)
df['size']=df['size'].astype(float)
```

```python
# NAN QIYMATLARNI ALOHIDA AJRATIB OLIB DATAFRAMEDAN O'CHIRIB TASHLADIK KEYIN ISHLATAMIZ
df_missing_price = df[df['price'].isna()]
df = df.dropna(subset=['price'])
df = df[df['price'] < 10_000_000]
df = df[df['price'] >= 1000]
df = df.drop(columns=['location'])
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
strafied_splits=StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
```

```python
for train_ix,test_ix in strafied_splits.split(df,df['district']):
    strat_train_set=df.iloc[train_ix]
    strat_test_set=df.iloc[test_ix]
```

```python






from sklearn.base import BaseEstimator,TransformerMixin
```

```python
room_ix,size_ix=1,2
class CombineAttributeAdder(BaseEstimator,TransformerMixin):
    def fit(self,X,y=None):
        return self
    def transform(self,X):
        size_per_room=X.iloc[:,size_ix]/X.iloc[:,room_ix]
        return np.c_[X,size_per_room]
```

```python
X_num=X.drop('district',axis=1)




# STANDARTSCALER
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# standart_scaler=StandardScaler()

# standart_scaler.fit(X_num)
# standart_scaler.transform(X_num)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline=Pipeline([
    ('attribs_adder',CombineAttributeAdder()),
    ('std_scaler',StandardScaler())
])
```

```python
num_attribs=list(X_num)
cat_attribs=['district']
full_pipeline=ColumnTransformer([
    ('num',num_pipeline,num_attribs),
    ('cat',OneHotEncoder(),cat_attribs)])
```

```python
X_prepared=full_pipeline.fit_transform(X)
X_prepared[:10]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR_model=LinearRegression()
LR_model.fit(X_prepared,y)
```

```python
test_data=X.sample(10)
test_label=y.loc[test_data.index]
```

```python
test_data_prepared=full_pipeline.transform(test_data)
```

```python
predicted_labels=LR_model.predict(test_data_prepared)
```

```python
pd.DataFrame({'bash':predicted_labels,'asl':test_label})
```

```python
from sklearn.metrics import mean_absolute_error

mae=mean_absolute_error(test_label,predicted_labels)
```

```python
print('MAE_LR=',mae)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(test_label,predicted_labels)
print('RMSE_LR=',np.sqrt(mse))
r2 = r2_score(test_label,predicted_labels)
print("R^2 Score:", r2)
```
