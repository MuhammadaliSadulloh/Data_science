```python
import pandas as pd

# Ma'lumotlarni yuklab olish
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population.csv", index_col=0)

# Aholi zichligini hisoblash
df['density'] = df['Population'] / df['Area (km²)']

# Aholi zichligi bo‘yicha kamayish tartibida saralash (eng kam zichlikdagilar oxirida bo‘ladi)
df_sorted = df.sort_values(by='density', ascending=True)

# Oxiridan 3-o‘rindagi davlatni olish
third_from_last = df_sorted.iloc[2]  # Oxirdan 3-o‘rin

# Maydonini olish
maydon = third_from_last['Area (km²)']

print(f"Aholi zichligi bo‘yicha oxiridan 3-o‘rindagi davlatning maydoni: {maydon} km²")
```
