```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.neural_network import MLPRegressor
%matplotlib inline

data = make_regression(n_samples=40,n_features=1,noise=10,random_state=42)

x_data,y_data=data[0],data[1]

def classic_program(x):return x*70
def weight_assigned_program(x,w):return w*x*70
w=1.4
regression=LinearRegression().fit(x_data,y_data)
mlp_regression=MLPRegressor(max_iter=2000).fit(x_data,y_data)

y_pred_klassik=[classic_program(x) for x in x_data]
y_pred_weight=[weight_assigned_program(x,w) for x in x_data]
y_pred_ML=regression.predict(x_data)
y_pred_DL=mlp_regression.predict(x_data)

plt.scatter(x_data,y_data,label='Haqiqiy data')
plt.plot(x_data,y_pred_klassik,'r',alpha=0.63,label='Klassik dastur')
plt.plot(x_data,y_pred_weight,'#FF6600',label='weight dastur')
plt.plot(x_data,y_pred_ML,color='#66FF00',label='ML')
plt.plot(x_data,y_pred_DL,color='#333300',alpha=0.49,label='DL')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
equation=f'{regression.coef_[0]:.3f}x+{regression.intercept_:.3f}'
plt.title(f'Klassik dastur: x*70 \nWeight dastur: {w}*x*70\nMachine Learning: {equation}')
plt.show()
```
